```python
#import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
#buat variable
x0=np.random.randn(1000) #ini merupakan ndarray

#buat plot
plt.plot(x0) #kalo ndarray dipanggilnya harus pake plt
```

```python
#buat plot pacf
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x0, ax=ax)
```

```python
#buat variable x1
x1=[0]
for i in range(1000):
  x=0.5*x1[-1]+0.1*np.random.randn()
  x1.append(x)

x1=np.array(x1)

#buat plot x1
plt.plot(x1)
```

```python
#buat plot pacf
fig, ax = plt.subplots(figsize=(10,5))
plot_pacf(x1, ax=ax)
```

```python
#buat iid noise
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(np.random.randn(1000), ax=ax)
```

```python
errors=0.1*np.random.randn(1000)
mal=[]
for i in range(1000):
  if i<=1:
    x=0.5*errors[i-1]+errors[i]
  else:
    x=errors[i]
  mal.append(x) #untuk menambahkan item diakhir list
mal=np.array(mal)

#buat plot
plt.plot(mal)
```

```python
#buat plot dgn acf
fig, ax = plt.subplots(figsize=(10,5))
plot_acf(mal,ax=ax)
```

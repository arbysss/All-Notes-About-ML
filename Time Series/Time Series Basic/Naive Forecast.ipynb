```python
# import MAPE, MAE, R2, dan MSE
from sklearn.metrics import mean_absolute_percentage_error, \
mean_absolute_error, r2_score, mean_squared_error
```

```python
# Download file SPY
!wget -nc https://lazyprogrammer.me/course_files/SPY.csv
```

```python
# buat dataframe dari file yang baru didownload di atas
df = pd.read_csv('SPY.csv', index_col='Date', parse_dates=True)

# tampilkan 5 data teratas dari dataframe di atas
df.head()
```

```python
# buat variable untuk menampung shift() dari close
df['ClosePrediction'] = df['Close'].shift(1) 
# shift() berguna untuk menambahkan 1 kolom baru (ClosePrediction) setelah (Close)

# display
df['ClosePrediction']
```

```python
# atur data actual dengan kolom Close
y_true = df.iloc[1:]['Close']

# atur data prediksi dengan kolom close prediksi
y_preds = df.iloc[1:]['ClosePrediction']
```

# Metrics

Main idea: rasakan bagaimana nilai2 tersebut saling berhubungan. Apakah baik? apakah buruk? jika R2 baik maka apakah MAE juga baik?

```python
# SSE
(y_true - y_preds).dot(y_true - y_preds)
```

```python
# MSE
mean_squared_error(y_true, y_preds)
```

```python
# MSE again (opsi ke-2 selain di atas)
# jangan takut untuk mengimplementasikannya sendiri
# ini sangat mudah (sangat baik untuk latihan)
(y_true - y_preds).dot(y_true - y_preds) / len(y_true)
```

```python
# RMSE
mean_squared_error(y_true, y_preds, squared=False)
```

```python
# RMSE (opsi ke-2 selain di atas)
np.sqrt((y_true - y_preds).dot(y_true - y_preds) / len(y_true))
```

```python
# MAE
mean_absolute_error(y_true, y_preds)
```

```python
# R2
r2_score(y_true, y_preds)
```

```python
# MAPE
mean_absolute_percentage_error(y_true, y_preds)
```

# sMAPE

```python
# sMAPE
# untuk sMAPE dapat menggunakan code berikut:

# buat fungsi
def smape(y_true,y_preds):
  numerator = np.abs(y_true - y_preds)
  denominator = (np.abs(y_true) + np.abs(y_preds)) / 2
  ratio = numerator/denominator
  return ratio.mean()

#display sMAPE
smape(y_true,y_preds)
```

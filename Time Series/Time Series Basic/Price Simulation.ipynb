```python
# import
import numpy as np
import matplotlib.pyplot as plt

# number of time steps (inisiasi jumlah waktu)
T = 1000

# initial price (inisiasi harga awal)
P0 = 10

# drift
mu = 0.001
```

```python
# last log price (log harga terakhir)
last_p = np.log(P0)

log_returns = np.zeros(T)
prices = np.zeros(T)
for t in range(T):
  # sample a log return (sampel pengembalian log)
  r = 0.01 * np.random.randn()

  # compute the new log price (hitung log harga baru)
  p = last_p + mu + r

  # store the return and price (simpan log kembali dan exponensial harga)
  log_returns[t] = r + mu
  prices[t] = np.exp(p)

  # assign last_p (menetapkan last_p)
  last_p = p
```

```python
# buat figure
plt.figure(figsize=(20, 8))

# buat plot dari prices
plt.plot(prices)

# buat judul figure
plt.title("Simulasi Harga")
```
